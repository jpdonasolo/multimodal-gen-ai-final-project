```python
import os
import sys
sys.path.append("SSV2A/")

from pathlib import Path
from typing import Literal, Union


import torch

from ssv2a.model.pipeline import Pipeline, image_to_audio
from setup import setup


IMG_FOLDER = "img"
MODEL: Union["yolo26x", "yolov8x-oiv7"] = "yolov8x-oiv7"
OUT_FOLDER = f"output/{MODEL}"
CONFIG = "models/ssv2a.json"
CHECKPOINT = "models/ssv2a.pth"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


if Path(os.getcwd()).stem == "src":
    os.chdir("..")
```

```python
setup(MODEL)
```

```python
images = [x for x in Path(IMG_FOLDER).glob("*")]
image_to_audio(
    images, 
    text="", 
    transcription="", 
    save_dir=OUT_FOLDER, 
    config=CONFIG,
    gen_remix=True, 
    gen_tracks=False, 
    emb_only=False,            
    pretrained=CHECKPOINT, 
    batch_size=64, 
    var_samples=64,
    shuffle_remix=True, 
    cycle_its=64, 
    cycle_samples=64,
    keep_data_cache=False, 
    duration=10, 
    seed=42, 
    device=DEVICE
)
```
